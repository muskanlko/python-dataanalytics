# graphs
1. univariate graphs: graphs that take a single variable
    - line plot-for growth/decay over a period
    - histogram- for distribution of data
    - density plot- for distribution of normalization from
    - box plot- for distribution of the data in quartiles
    - pie chart- for distribution of the datra in persentage
    - bar chart- for discrete data comparison
2. bivariate graphs:graph that take two variable
    - scatter plot- relation b/w two variable
    - heat map- relation b/w two varible in matrix from
    - violin plot - relation b/w variable in density from
    - pair plot- relation b/w variable in grid form
3. multivariate graphs: graphs that take more than two
    - bubble plot- relation b/w 2 var with size
    - 3D scatter plot- relation b/w 3 var on x,y,z axis
    - stacked bar chart- relation b/w 3 variable in bar form 

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
x=np.random.normal(size=1000)
plt.hist(x,bins=np.arange(-3,3,.25),edgecolor='black')
plt.show()
```

```python
sns.histplot(x,bins=np.arange(-3,3,.25),kde=True)
```

```python
px.histogram(x,nbins=20)
```

```python
# box plot 
x=np.random.normal(size=1000)
plt.boxplot(x)
plt.show()
```

```python
sns.boxplot(x)# boxplot in alternative of histrogram
```

```python
px.box(x)
```

```python
#pie chart
rainfall=[20,99,12,56,26]
labels=['Jan','Feb','Mar','Apr','May']
plt.pie(
    rainfall,
    autopct="%.2f%%",
    labels=labels,
)
plt.show()
```

```python
# area chart
x=np.arange(20)
y=np.random.randint(1,20,20)
plt.fill_between(x,y)
plt.show()
```

```python
px.area(x=x,y=y)
```

```python
years=np.arange(2010,2020,dtype=int)
sales=np.random.randint(1000,5000,10)
plt.bar(years,sales)
```

```python
sns.barplot(x=years,y=sales)
```

```python
px.bar_polar(r=years,theta=sales)
```

```python
x=np.random.normal(size=1000)
y=np.random.normal(size=1000)
plt.scatter(x,y,alpha=.4,edgecolor='black')
```

```python
sns.scatterplot(x=x,y=y)
```

```python
px.scatter(x=x,y=y)
```

```python
sns.heatmap(np.random.rand(10,10))
```

```python
plt.contour(np.random.rand(10,10))
```

```python
px.density_heatmap(x=x,y=y)
```

```python
#violin plot
x=np.random.normal(size=1000)
sns.violinplot(x)
```

```python
sns.swarmplot(x)
```

```python
px.violin(x)
```

```python
# bubble chart
x=np.random.normal(size=300)
y=np.random.normal(size=300)
z=np.random.randint(10,200,300)
plt.scatter(x,y,s=z,alpha=.4,c=z)
```

```python
px.scatter(x=x,y=y,size=z,color=z)
```
