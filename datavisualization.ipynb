```python
penguins=sns.load_dataset("penguins")
penguins
```

```python
sns.countplot(data=penguins,x="species")
```

```python
sns.countplot(data=penguins,x="island",hue="species")
```

```python
sns.countplot(data=penguins,y="island",hue="species",fill=False)
```

```python
sns.histplot(data=penguins,x="body_mass_g",bins=10,kde=True)
```

```python
sns.histplot(data=penguins,x="flipper_length_mm",bins=10,kde=True)
```

```python
sns.swarmplot(
    data=penguins,
    x="island",
    y="body_mass_g",
    hue="species"
    )
```

```python
sns.relplot(
    data=penguins,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species",
    col="species"
)
```

```python
sns.relplot(
    data=penguins,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species",
    col="sex"
)
```

```python
sns.relplot(
    data=penguins,
    x="flipper_length_mm",
    y="bill_depth_mm",
    hue="species",
    col="species"
)
```

```python
px.scatter_3d(
    data_frame=penguins,
    x="flipper_length_mm",
    y="bill_length_mm",
    z="body_mass_g",
    color="species"
)
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
sns.countplot(data=tips,x="sex",hue="time")
```

```python
sns.scatterplot(data=tips,
                x="total_bill",
                y="tip",
                hue="sex")
```

```python
sns.catplot(data=tips,
            x="day",
            y="total_bill",
            hue="sex")
```

```python
sns.catplot(data=tips,
            x="day",
            hue="smoker",
            kind="count")
```

```python
sns.catplot(data=tips,
            x="time",
            y="tip",
            hue="sex",
            kind="box")
```

```python
# sunburst
px.sunburst(
    data_frame=tips,
    path=["day","time","sex","smoker"],
    values="tip"
)
```

```python
# sunburst
px.treemap(
    data_frame=tips,
    path=["day","time","sex","smoker"],
    values="tip"
)
```
